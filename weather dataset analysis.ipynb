##### data preprocessing

```python
# to get Data types of columns in Dataframe
print(df. dtypes)

# converting all non-numeric columns to numeric columns

# enc.fit() creates the corresponding integer values.
# enc.transform() applies the encoding to the df values.
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()

# Specify the dtype to extract with the include parameter.
non_numeric_data = df.select_dtypes(include=object)
non_numeric_columns = non_numeric_data.columns.values
print("\nnon_numeric_columns in the dataframe are as follows :\n", non_numeric_columns)
for i in range(len(non_numeric_columns)):
    col = non_numeric_columns[i]
#     print(df[col])
    
    enc.fit(df[col])
    df[col] = enc.transform(df[col])
#     print(df[col])

# to ensure Data types of columns in Dataframe is numeric
print(df. dtypes)
```

```python
df['RainTomorrow'] = df['RainTomorrow'].replace(['Yes'],1)
df['RainTomorrow'] = df['RainTomorrow'].replace(['No'],0)
df['RainTomorrow']
```

##### train-test split

```python
from sklearn.model_selection import train_test_split

X = df.drop(['RainTomorrow'], axis = 1)
y = df['RainTomorrow']

training_data, testing_data, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")
```

```python
# ValueError: Input contains NaN, infinity or a value too large for dtype('float64').. the below two blocks to handle this error
# https://stackabuse.com/k-nearest-neighbors-algorithm-in-python-and-scikit-learn/
# https://researchdatapod.com/python-valueerror-input-contains-nan-infinity-or-a-value-too-large-for-dtype-float64/
```

```python
# Feature Scaling -  to scale the features so that all of them can be uniformly evaluated.

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(training_data)

training_data = scaler.transform(training_data)
testing_data = scaler.transform(testing_data)
training_data = pd.DataFrame(training_data)
testing_data = pd.DataFrame(testing_data)
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=0)
imputer = imp_mean.fit(training_data)
training_data = imputer.transform(training_data)

imp_mean = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=0)
imputer = imp_mean.fit(testing_data)
testing_data = imputer.transform(testing_data)
```

##### KNN classifier

```python
# KNN classifier

from datetime import datetime
now = datetime.now()
start_time = now.strftime("%H:%M:%S")
print("start time =", start_time)

accList = []
# Import necessary modules
from sklearn.neighbors import KNeighborsClassifier
for i in range(40,45):
    knn = KNeighborsClassifier(n_neighbors=i)

    knn.fit(training_data, y_train)

    # Predict on dataset which model has not seen before
    y_pred = knn.predict(testing_data)
#     print(y_pred)

    import numpy as np
    # converting series to ndarray
    y_test = np.asarray(y_test)

    # confusion matrix
    from sklearn.metrics import confusion_matrix
    tp_tn, fp_fn = confusion_matrix(y_test, y_pred)
#     print(confusion_matrix(y_test, y_pred))
    print(f"true positive : {list(tp_tn)[0]}\n true negative : {list(tp_tn)[1]}\n false positive : {list(fp_fn)[0]}\n false negative : {list(fp_fn)[1]}\n")

    # error evaluation methods
    # calculating accuracy

    from sklearn.metrics import classification_report
    from sklearn.metrics import accuracy_score

    #Check precision, recall, f1-score
    print(classification_report(y_test, y_pred) )

    print("accuracy_score",accuracy_score(y_test, y_pred))
    print("is with k = ",i)
    accList.append(accuracy_score(y_test, y_pred))
    print("=======================================================================================")

from statistics import mean
print("mean of the accuracy is")
mean(accList)


now = datetime.now()
end_time = now.strftime("%H:%M:%S")
print("end time =", end_time)

# computing time complexity
time_1 = datetime.strptime(start_time,"%H:%M:%S")
time_2 = datetime.strptime(end_time,"%H:%M:%S")

time_interval = time_2 - time_1
print(time_interval)
 
```
